```python
# Median value of horsepower
median = df.horsepower.median()

print(f"The following is the median value of horsepower column:\n " + "-"*100 +  f"\n {median}")
```

```python
# Most frequent value of the horsepower column
mode = df['horsepower'].mode()[0]

print(f"The following is the mode(most frequent) value of horsepower column:\n " + "-"*100 +  f"\n {mode}")
```

```python
# Handle missing data (replace missings horsepower values with mode)
print("Missing values in the column 'horsepower': \n ")
print(df['horsepower'].isna().sum())

print("-"*100)
df['horsepower'] = df['horsepower'].fillna(mode)
print("After replacement")
print("-"*100)
print(df['horsepower'].isna().sum())
```

```python
median = df.horsepower.median()

print(f"The following is the new median value of horsepower column after missing values are replaced:\n " + "-"*100 +  f"\n {median}")
```

## Question 7
1. Select all the cars from Europe
2. Select only columns vehicle_weight and model_year
3. Select the first 7 values
4. Get the underlying NumPy array. Let's call it X.
5. Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
6. Invert XTX.
7. Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].
8. Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
9. What's the sum of all the elements of the result?

```python
# Sum of weights
df_cars_from_asia = df.loc[df['origin'] == 'Asia']

df_cars_from_asia = df_cars_from_asia.reset_index(drop=True)
df_cars_from_asia[['vehicle_weight', 'model_year']].head(7)
```

```python
X = np.array(df_cars_from_asia[['vehicle_weight', 'model_year']].head(7))
X.shape
```

$$
w = (X^T X)^{-1} X^T y
$$

```python
# X transpose 
XTX = X.T @ X


# Invert XTX
XTX_inv = np.linalg.inv(XTX)

# array
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])


# weight
w = XTX_inv @ X.T @ y
print(w[0] + w[1])
# sum(w)
```
